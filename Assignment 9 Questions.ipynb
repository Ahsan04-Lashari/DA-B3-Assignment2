### Question 2:

How does time spent on social media compare to time spent on education among teens?

- Plot side-by-side histograms of Time on Social Media and Time on Education
- Find and compare the median time for both.
- Write simple comments to explain what the data shows

```python


# Step 2: Extract relevant columns
social_time = df['Time_on_Social_Media']
education_time = df['Time_on_Education']

# Step 3: Plot side-by-side histograms
plt.figure(figsize=(10, 6))
plt.hist(social_time,  label='Social Media', color='blue')
plt.hist(education_time, label='Education', color='green')
plt.title('Time Spent on Social Media vs Education (Teens)')
plt.xlabel('Hours')
plt.ylabel('Number of Teens')
plt.legend()
plt.grid(True)
plt.show()

# Step 4: Calculate medians
social_median = social_time.median()
education_median = education_time.median()

# Step 5: Print the medians
print(f"Median Time on Social Media: {social_median} hours")
print(f"Median Time on Education: {education_median} hours")

# Step 6: Comments on the data
if social_median > education_median:
    print("Teens generally spend more time on social media than on education.")
elif social_median < education_median:
    print("Teens generally spend more time on education than on social media.")
else:
    print("Teens spend equal time on social media and education on average.")
```

### Question 3

- How can we detect and identify outliers in Daily Usage Hours, Exercise Hours, and Screen Time Before Bed among teens using the IQR method?
- Display the records corresponding to these outliers using boxplot.



```python
# List of columns to check for outliers
columns_to_check = ['Daily_Usage_Hours', 'Exercise_Hours', 'Screen_Time_Before_Bed']

# Dictionary to store outliers
outliers = {}

# Detect outliers using IQR method
for column in columns_to_check:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter outliers
    outlier_rows = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    outliers[column] = outlier_rows
    
    print(f"\nOutliers in '{column}':")
    print(outlier_rows[[column]])

# Display boxplots for visualizing outliers
plt.figure(figsize=(12, 6))
sns.boxplot(df=df[columns_to_check], orient='h')
plt.title("Boxplots Showing Outliers in Teen Activity Data")
plt.xlabel("Hours")
plt.show()
```

### Question 5

How does the average addiction level vary across different age groups? Show with line graph

```python
avg_addiction_by_age = df.groupby('Age')['Addiction_Level'].mean().reset_index()
plt.figure(figsize=(8, 5))
plt.plot(avg_addiction_by_age['Age'], avg_addiction_by_age['Addiction_Level'],
         marker='o', linestyle='-', color='darkred')
plt.title("Average Addiction Level Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Average Addiction Level")
plt.grid(True)
plt.tight_layout()
plt.show()
```
